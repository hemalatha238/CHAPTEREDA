```python
import pandas as pd

time_series = [11, 11.30 , 12 , 12.30 , 13 , 13.30]
print("Before DataFrame :", time_series)

mydata = pd.Series(time_series)
print("----------------------")
print("After Dataframe :")
print(mydata)
```

```python
import pandas as pd
import numpy as np
marks = np.array([86 , 75 , 67 , 78 , 98])

dataFrame = pd.Series(marks , index = ["Math" , "Scie" , "Eng" , "Hindi" , "Telagu"])
print(dataFrame)
```
