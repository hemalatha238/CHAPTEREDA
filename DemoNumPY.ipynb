```python
import numpy
num_array = numpy.array([10 , 20 , 30 , 40], dtype = "int64")
#num_array = numpy.array([10 , 20 , 30 , 40, "City"], dtype = "int64")
print(num_array)
print(type(num_array))
print(num_array.dtype) #using dtype we are able to see data type of array collection
```

```python
#num_array = numpy.array([10 , 20 , 30 , 40 , "Name"], dtype = "float64")
num_array = numpy.array([10 , 20 , 30 , 40 , "Name"], dtype = "U")
#dtype = U  (means string data type)
print(num_array)
```

```python
import numpy as np
array = np.array([1 ,2 ,3, 4 ,5 ,6 ,7 ,8])
print(array[0])
print(array[1])
print(array[2:])
print(array[3:7])
print(array.ndim)
```

```python
a = np.array([[1 ,2 ,3 ], [11, 22 , 33], [111, 222 ,333 ]])
print(a.shape)
print(a[0][0])
print(a[0][1])
print(a[1][0])
print(a[2][0])
print(a[1][2])
a[0][0] = 99
print(a)
print(a.ndim)
```

```python
a = np.array([[[1 ,2 ,3 ], [11, 22 , 33], [111, 222 ,333 ]]])
print(a.ndim)
```
