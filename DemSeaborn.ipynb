```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
std_data = {
    "roll_no" : np.arange(101 , 111),
    "name" : ["Sonu" , "Monu" , "Golu" , "Molu", "Sweety" , "Rani" , "rutu" , "Nitu" , "Meetu", "Hari"],
    "gender" :["F" , "M" , "F" , "F" , "M" , "F" , "M" , "F" , "F" , "M" ],
    "age" : np.random.randint(20 , 25 , 10),
    "study_hours" : np.random.randint(2 , 6 , 10),
    "math_marks" : np.random.randint(50 , 100 , 10),
    "sci_marks" : np.random.randint(50 , 100 , 10),
    "eng_marks" : np.random.randint(50 , 100 , 10),
}
data_frame = pd.DataFrame(std_data)
#print(data_frame)
```

```python
sns.violinplot(data = data_frame , x ="roll_no" , y = "eng_marks")
plt.show()
```

```python
sns.displot(data=data_frame, x="roll_no", col="eng_marks", kde=True)
plt.show()
```

```python
sns.scatterplot(x="study_hours" , y="math_marks" , data = data_frame)
plt.title("Study Hours vs Math Score")
plt.show()
```

```python
sns.lineplot(x="roll_no" , y ="study_hours",data = data_frame)
plt.title("Study Hours by Students roll no")
plt.show()
```

```python
# Import seaborn
import seaborn as sns
import pandas as pd

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
data_frame = pd.DataFrame(tips)

```

```python
# Create a visualization
sns.relplot(
    data= data_frame,
    x="total_bill", y="tip", col="time",
    hue="smoker" , style="smoker", size="size",
)
```
