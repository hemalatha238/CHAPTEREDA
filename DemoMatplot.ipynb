```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [100 , 50 , 30 , 70 , 90]
plt.figure(figsize=(5, 2.5)) # resize the graph
plt.plot(x , y ,  linestyle = "dashed" ,color = "red"  , marker = 'o' ,)
plt.title("Ploting a Line")
plt.xlabel("X - axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([2020 , 2021 , 2022 , 2023 , 2024])
sale_per_year =np.array([25000 , 30000, 4500 , 12000 , 40000])

plt.figure(figsize=(10, 5)) # resize the graph
plt.bar(years , sale_per_year , color = "orange")
plt.title("Sales per Year")
plt.xlabel("Years")
plt.ylabel("Sales in RS")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

gender = np.array([100 , 50 , 30])
mylabel = ["Female" , "Male" , "Others"]
mycolour = ["hotpink" , "blue" , "red"]
popup = [0, 0.5 , 0]

plt.pie(gender , labels = mylabel , colors= mycolour , startangle= 90 , explode=popup)
plt.legend()
plt.title("KIET Students Gender")
plt.show()


#Fruits  5
#City - Jaipur , Nagpur ,  5
#states
#vegitables
#electronics
#Sales of last five year
#countries  
```
